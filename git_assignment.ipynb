reate a Git working directory, with the following branches:


               Develop

               F1

               f2

```python
-# Create a new directory
mkdir git-working-dir
cd git-working-dir

# Initialize Git repository
git init

# Create an initial file
echo "Initial content" > README.md

# Add and commit the file
git add README.md
git commit -m "Initial commit"

# Create the required branches
git branch Develop
git branch F1
git branch f2

# List all branches
git branch
```

2. In the master branch, commit main.txt file

```python
# Switch to master branch
git checkout master

# Create the file
echo "This is the main branch file" > main.txt

# Stage the file
git add main.txt

# Commit the changes
git commit -m "Add main.txt to master branch"
```

3. Put develop.txt in develop branch, f1.txt and f2.txt in f1 and f2 respectively

```python
# Switch to develop branch
git checkout develop
echo "This is develop branch" > develop.txt
git add develop.txt
git commit -m "Add develop.txt"

# Switch to f1 branch
git checkout f1
echo "This is f1 branch" > f1.txt
git add f1.txt
git commit -m "Add f1.txt"

# Switch to f2 branch
git checkout f2
echo "This is f2 branch" > f2.txt
git add f2.txt
git commit -m "Add f2.txt"
```

4. Push all these branches to GitHub

```python
# Add your GitHub repository as the remote
git remote add origin https://github.com/<your-username>/<your-repository>.git

# Push the master branch
git push -u origin master

# Push the develop branch
git push -u origin develop

# Push the f1 branch
git push -u origin f1

# Push the f2 branch
git push -u origin f2
```

5. On local delete f2 branch

```python
# Switch to another branch (for example, master)
git checkout master

# Delete the local f2 branch
git branch -d f2
```

6. Delete the same branch on GitHub as well

```python
git push origin --delete f2
```

question answer 4

```python
# 1. On master branch, create master.txt, stage and commit
git checkout master
echo "Master file" > master.txt
git add master.txt
git commit -m "Add master.txt"

# 2. Create three branches
git branch public1
git branch public2
git branch private

# 3. Add public1.txt on public1 branch, stage and commit
git checkout public1
echo "Public 1 file" > public1.txt
git add public1.txt
git commit -m "Add public1.txt"

# 4. Merge public1 into master
git checkout master
git merge public1

# 5. Merge public2 into master
git merge public2

# 6. Edit master.txt on private branch, stage and commit
git checkout private
echo "Updated content in private branch" >> master.txt
git add master.txt
git commit -m "Update master.txt in private"

# 7. Update public1 and public2 with new code from private
git checkout public1
git merge private

git checkout public2
git merge private

# 8. Update master with new code from private
git checkout master
git merge private

# 9. Finally update private with the latest master code
git checkout private
git merge master
```
